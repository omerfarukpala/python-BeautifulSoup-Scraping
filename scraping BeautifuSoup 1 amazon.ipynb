```python
from bs4 import BeautifulSoup
import requests
import time
```

```python
request = requests.get("https://www.amazon.com.tr/gp/bestsellers/electronics/13709880031/ref=zg_bs_nav_1_electronics")
request
# verileri aldığımızda 200 döner
```

verileri çekebildiğimizde 200 döner 

```python
soup = BeautifulSoup(request.content, "lxml")
```

```python
# soup.prettify()
```

prettify() daha okunalı hale getirir.

```python
ürünler = soup.find_all("div",attrs={"class":"_cDEzb_grid-column_2hIsc"})
# ürünler
```

```python
header = {'User-Agent': "use your own user agent"}
```

network içinden user agent ı sözlük olarak girmeliyiz.

```python
for ürün in ürünler:
    ürün_linkleri = ürün.find_all("div",attrs={"class":"zg-grid-general-faceout"})
    for i in ürün_linkleri:
        link = i.find("div",attrs={"class":"p13n-sc-uncoverable-faceout"})
        link_devam = link.a.get("href")
        # evet sitede gördüğümüz linkleri aldık ama slashtan öncesi yani "https://www.amazon.com.tr" olmadığı için limk olarak içine giremiyoruz bu sebeple yazacağımız kodla başına ekliyoruz. 
        link_basi = "https://www.amazon.com.tr"
        link_tamami = link_basi+link_devam
        # print(link_tamami)
        
        detay = requests.get(link_tamami, headers = header)
        # print(detay)   # evet artık linkleri çekebiliyoruz
        detay_soup = BeautifulSoup(detay.content, "lxml")
        # BeautifulSoup() ile parçalıyoruz

        teknik_ayrıntılar = detay_soup.find_all("table", attrs={"class":"prodDetTable" ,"role": "presentation"})
        # print(teknik_ayrıntılar)
        for teknik in teknik_ayrıntılar:
                detaylar = teknik.find_all("tr")
                for i in detaylar:
                    try:
                        etiket = i.find("th", attrs={"class": "prodDetSectionEntry"}).get_text(strip=True, separator=" ")
                        deger = i.find("td", attrs={"class": "prodDetAttrValue"}).get_text(strip=True, separator=" ")
                        print(f"{etiket} = {deger}")
                    except:
                        print("---------")
```
